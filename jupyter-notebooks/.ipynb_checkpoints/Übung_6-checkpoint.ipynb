```r
p=0.5 # Wahrscheinlichkeit für Mädchengeburten liegt bei 50%
n=1000 # Neugeborene
x_all=530:560 # Wertebereich für genau x Mädchengeburten

x_all #print
```

```r
calculate_pbinom_multiple_x <- function(x_all,n,p){
    p_exact_all <- sapply(x_all, function(x) binom.test(x, n, p, alternative="greater")$p.value)
}


p_exact_all = calculate_pbinom_multiple_x(x_all,n,p) 
p_exact_all #print
```

```r
calculate_approx_norml_multiple_x <- function(x_all,n,p){
    z = ((x_all)-n*p)/sqrt(n*p*(1-p)) 
    p_approx_all = 1-pnorm(z)
}

p_approx_all = calculate_approx_norml_multiple_x(x_all,n,p) 
p_approx_all # print
```

```r
calculate_approx_norml_corrected_multiple_x <- function(x_all,n,p){
    z = ((x_all-0.5)-n*p)/sqrt(n*p*(1-p)) 
    p_approx_corrected_all = 1-pnorm(z)
}

p_approx_corrected_all = calculate_approx_norml_corrected_multiple_x(x_all,n,p) 
p_approx_corrected_all
```

```r
plot(x_all,p_exact_all, type="h", lwd=2, col="red", ylab="")
lines(x_all,p_approx_all, type="l", lwd=2, col="black", ylab="")
lines(x_all,p_approx_corrected_all, type="l", lwd=2, col="blue", ylab="")

legend(545, 0.03, legend=c("Exact Values", "Approximated Values", "Corrected Approximated Values"), col=c("red", "black","blue"), lty=1:1, cex=0.8)
```
