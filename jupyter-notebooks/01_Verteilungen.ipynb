# Wichtige Verteilungen

## Poisson-Verteilung (für Zähldaten z.B. Supermarktwarteschlangen oder radioaktiver Zerfall

```r
x = 1:24
lamda = 8 # Intensität = Erwartungswert = Varianz

y = dpois(x,lamda)
plot(x,y, type = "h", col = "red")
```

## Geometrische Verteilung (Erwarten auf Erfolg z.B. Bernoulli-Versuche mit p -> Wie lange muss ich den Versuch durchführen, bis ich Erfolg habe?)


```r
x = 0:10 # Anzahl der Misserfolge, bis zum ersten Erfolg
prob = 0.2 # Erfolgswahrscheinlichkeit, bei einem Bernoulli-Versuch

y = dgeom(x,prob)
plot(x,y, type = "h", col = "red")
```

## Negative Binomialverteilung (Verallgemeinerung der Geometrischen Vertilung z.B. Wie lange muss ich den Versuch durchführen, bis ich n Erfolge habe?)

```r
x = 0:10
prob = 0.2
size = 3

y = dnbinom(x,size,prob)
plot(x,y, type = "h", col = "red")
```

## Hypergeometrische Verteilung 
(gegeben: Untermenge einer Gesamtmenge -> Stichprobe. Wie viele Elemente der Untermenge habe ich in der Stichprobe?)
Z.B. Wahlbefragung -> Wie groß ist der Anteil der Parteiangehörigen von einer Befragung an der Gesamtpopulation?


```r
x = 0:10 # Vektor der Häufigkeiten der markierten gezogenen kugeln
k = 20 # gezogene Kugeln (ohne zurücklegen!)
m = 30 # Anzahl der Markierten
n = 10 # Anzahl der Nicht-Markierten

y = dhyper(x,n,m,k)
plot(x,y, type = "h", col = "red")
```

## Normalvertielung 
Wichtig: Zweiter Wert ist nicht die Varianz, sondern die Standardabweichung

```r
x = seq(-3,3,by=0.05)
mean = 0
sd = 1

y = dnorm(x, mean=mean, sd=sd)
plot(x,y,type="l",col="red")
```

### Beispiel Mädchengeburten mit Binomialverteilung und Normalverteilungsapproximation

```r
# Realwert bei Binomialvertielung bei Bsp Mädchengeburten
n = 1000
p = 0.5
k = 530

P = 1-pbinom(k-1,n,p)
print(P)
```

```r
# Normalvertielungsapproximation für Bsp Mädchengeburten
z = (k-n*p)/sqrt(n*p*(1-p))
P = 1-pnorm(z)
print(P)

# -> Mit Fehlerkorrektur
z = ((k-0.5)-n*p)/sqrt(n*p*(1-p))
P = 1-pnorm(z)
print(P)
```

## Uniforme Verteilung

```r
x = seq (-3,3, by =0.05)
y = dunif(x, -3,3)

plot(x,y, type = "l", lwd=2, col="red")
```

## Exponentialverteilung
zB Wartezeiten, Überlebenszeiten
Aufpassen, ob man eropäische (lamda), oder amerikanische Bezeichnung nimmt (1/lamda) <br>

Parameter: rate

```r
x = seq(0,3,by=0.05)
y = dexp(x,1) 
z = dexp(x,2)

plot(x,y, type = "l", lwd=2, col="red")
lines(z,y, type = "l", lwd=2, col="blue")
```

## Gammaverteilung
Exponentialvertielung und Chi-Quadratverteilung sind Sonderfälle <br>

parameter: rate, alpha

```r
x = seq(0,3,by=0.05)
y = dgamma(x,1,2)
z = dgamma(x,2,2)

plot(x,y, type = "l", lwd=2, col="red", ylim=c(0,2))
par(new=TRUE)
plot(x,z, type = "l", lwd=2, col="blue", ylim=c(0,2))
```

## Chi-Qadrat-Verteilung
Summe aus Quadranten unabhängiger Freiheitsgrade

```r
x = seq(0,4,by=0.01)
n = 1 # anzahl der freiheitsgrade

alpha = n/2
y = dchisq(x,n)
# z = dgamma(x, alpha, 0.5) # -> Äquivalent

plot(x,y, type = "l", lwd=2, col="blue", ylim=c(0,1))
```

## t-Verteilung
Risikomodellierung, um Abweichungen von Verteilungen zu modellieren
Konvergiert bei großem n gegen die Normalverteilung

```r
x = seq(-3,3,by=0.01)
n = 1 # anzahl der freiheitsgrade

y = dt(x,n)

plot(x,y, type = "l", lwd=2, col="blue", ylim=c(0,0.5))
```

## F-Verteilung
Vergelich zweier Grupppen von Daten, zB Frauen und Männerergebnisse bei Prüfungsergebnisse

```r
x = seq(0,4,by=0.01)
n = 3 # anzahl der freiheitsgrade zB bei Frauen
m = 1 # anzahl der anderen freiheitsgrade zB bei Männern

y = df(x,n,m)

plot(x,y, type = "l", lwd=2, col="blue", ylim=c(0,1))
```
